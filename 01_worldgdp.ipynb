```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# Chrome 드라이버 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 헤드리스 모드 (브라우저 창을 보이지 않게)
options.add_argument('--no-sandbox')  # 추가적인 옵션
options.add_argument('--disable-dev-shm-usage')  # 리소스 사용 최적화
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 웹 페이지 열기
url = 'https://mfront.homeplus.co.kr/list?categoryDepth=2&categoryId=200060'
driver.get(url)

# 특정 요소가 로드될 때까지 대기
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, '#site-wrapper > div.content-container.filterTopLine.css-1n2pp32'))
)

# 페이지 소스 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# CSS 선택자 사용
css_selector = '#site-wrapper > div.content-container.filterTopLine.css-1n2pp32 > div.category-container-wrap > div.category-prod-list-wrap > div.infinite-scroll-component__outerdiv > div > div.unitItemWrap.typeCard.css-dyb522'
gdp_list = soup.select_one(css_selector)

if gdp_list:
    # 데이터 추출
    rows = gdp_list.find_all('div', class_='detailInfoWrap')
    for row in rows:
        cols = row.find_all('div', class_='detailInfo')
        print([x.text.strip() for x in cols])
else:
    print("gdp_list를 찾을 수 없습니다.")

# 브라우저 종료
driver.quit()
```

```python
import asyncio
import nest_asyncio
from pyppeteer import launch
from bs4 import BeautifulSoup

# 중첩된 이벤트 루프 허용
nest_asyncio.apply()

async def main():
    # 브라우저 실행
    browser = await launch(headless=True, executablePath=r'C:\Users\samsung\AppData\Local\pyppeteer\pyppeteer\local-chromium\132.0.6783.0\chrome-win\chrome.exe')
    page = await browser.newPage()
    
    # 웹 페이지 열기
    url = 'https://mfront.homeplus.co.kr/list?categoryDepth=2&categoryId=200060'
    await page.goto(url, {'waitUntil': 'networkidle2'})  # 네트워크 안정화 대기

    # 특정 요소가 로드될 때까지 대기
    await page.waitForSelector('div.unitItemWrap.typeCard')  # 필요한 요소 대기

    # 페이지 소스 가져오기
    html = await page.content()

    # BeautifulSoup으로 파싱
    soup = BeautifulSoup(html, 'html.parser')

    # CSS 선택자 사용
    gdp_list = soup.select_one('div.unitItemWrap.typeCard')

    if gdp_list:
        # 데이터 추출
        rows = gdp_list.find_all('div', class_='detailInfoWrap')
        for row in rows:
            cols = row.find_all('div', class_='detailInfo')
            print([x.text.strip() for x in cols])
    else:
        print("gdp_list를 찾을 수 없습니다.")

    # 브라우저 종료
    await browser.close()

asyncio.run(main())
```

```python
import requests
import json
url = 'https://mfront.homeplus.co.kr/category/item.json?categoryDepth=2&categoryId=200060&page=1&perPage=20&sort=RANK'
res = requests.get(url)
# JSON 형식의 문자열을 로드
j = json.loads(res.text)
products = j['data']['dataList']
for product in products:
    print(product['itemNm'],product['salePrice'])
```
