```python
import pandas as pd

# 열 2개를 가진 데이터프레임 생성
# 행 단위로 데이터 입력
df = pd.DataFrame([[1,2],[2,3],[3,4]])
df
```

```python
# 열 이름 지정
df = pd.DataFrame([[1,2],[2,3],[3,4]], columns=['ColA','ColB'])
df
```

```python
# 인덱스 지정
df = pd.DataFrame([[1,2],[2,3],[3,4]], 
                  columns=['ColA','ColB'], 
                  index=['Row1','Row2','Row3'])
df
```

```python
data = [
    {'ColA': 1, 'ColB': 'A'},
    {'ColA': 2, 'ColB': 'B'},
    {'ColA': 3, 'ColB': 'C'},
]
df = pd.DataFrame(data)
df.to_csv('my.csv')
```
